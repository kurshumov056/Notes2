```python
'''    
                                                    3. Catalogue 

Create a class Catalogue. The __init__ method should acccept the name of the catalogue (string). Each catalogue should also have an attribute called products, an empty list.
The class should also have three more methods:

    - add_product(product_name: str) - adds the product to the products list.
    - get_by_letter(first_letter: str) - returns a list containing only the produts that start with the given letter
    - __repr__ - returns the catalogue info in the following format:
    "Items in the (name) catalogue:
    {item1}
    {item2}
    ...
    (itemN):"

    
    The items should be sorted alphabetically in ascending oreder.

Example:


            Test Code                                   Output

    catologue = Catalogue("Furniture")                  [Chair", "Carpet"]
    calogue.add_product("Sofa")                         Items in the furniture catalogue:
    calogue.add_product("Mirror")                       Carpet
    calogue.add_product("Desk")                         Chair
    calogue.add_product("Chair")                        Desk
    calogue.add_product("Carpet")                       Mirror
    print(catalogue.get_by_letter("C"))                 Sofa
    print(catalogue)                      
'''

class Catalogue:

    def __init__(self, catalogue):
        self.catalogue = catalogue
        self.lst = []
        self.string_result = ""
        self.result = ""
        self.result = "Items in the {} Catalogue are: ".format(self.catalogue)

    

    def add_product(self, product):
        self.product = product
        self.lst.append(self.product)

    def get_by_letter(self, letter):
        self.letter = letter 
        self.lst_by_letter = []
        
        for x in self.lst:
            if str(x)[0] == self.letter:
                self.lst_by_letter += [x]
        return "Catalogue starting with letter '{}': {}".format(self.letter,self.lst_by_letter)
    
    def __repr__(self):

        for element in self.lst:
            self.result += "{} \n".format(element)
        
        return self.result + self.string_result



    


products = Catalogue('Home Furniture')

products.add_product("Sofa")                      
products.add_product("Mirror")                     
products.add_product("Desk")                         
products.add_product("Chair")                      
products.add_product("Carpet")
 
print(products.get_by_letter("C"))
print(products)
 
```
